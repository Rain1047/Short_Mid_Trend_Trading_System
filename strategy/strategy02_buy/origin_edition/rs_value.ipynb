```python
## 52 week high and low
one_year_high = hist.tail(365).close.max()
one_year_low = hist.tail(365).close.min()
## 60 day high and low
sixty_day_high = hist.tail(60).close.max()
sixty_day_low = hist.tail(60).close.min()
# 创建单个ticker的row
temp = pd.DataFrame()
temp['symbol'] = 'symbol'
temp['datetime'] = hist.tail(1).datetime
temp.symbol = ticker

# origin editon
# condition1: close > MA50 > MA150 > MA200 
# condition2: close > max(52week_low * 1.25, 52week_high*0.75)
# condition3: rs
temp['con1'] = 0
temp['con2'] = 0
if hist.iloc[-1].close > hist.iloc[-1].MA50 > hist.iloc[-1].MA150 > hist.iloc[-1].MA200:
    temp['con1'] = 1
# improve editon
# condition1: close>20 MA10>MA50 MA20>MA200 MA50>MA200
# condition2: 
if hist.iloc[-1].MA150 > hist.iloc[-1].MA20 and hist.iloc[-1].MA10 > hist.iloc[-1].MA50 and hist.iloc[-1].MA20 > hist.iloc[-1].MA200 and hist.iloc[-1].MA50 > hist.iloc[-1].MA200:
    temp['con2'] = 1

# 52 week high and low
# 注：为1就是好的
if hist.iloc[-1].close >= one_year_high *0.75:
    temp['52-week-high-zone'] = 1
else: 
    temp['52-week-high-zone'] = 0
if hist.iloc[-1].close >= one_year_low * 1.25:
    temp['52-week-low-zone'] = 1
else:
    temp['52-week-low-zone'] = 0
if hist.iloc[-1].close > sixty_day_high *0.9:
    temp['60-day-high-zone'] = 1
else:
    temp['60-day-high-zone'] = 0
if hist.iloc[-1].close > sixty_day_low * 1.1:
    temp['60-day-low-zone'] = 1
else:
    temp['60-day-low-zone'] = 0
```
