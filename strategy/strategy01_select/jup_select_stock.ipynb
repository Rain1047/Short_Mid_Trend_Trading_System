```python
# Pandas
import pandas as pd
import talib as ta
import numpy as np
# from get_score import get_score_ch
from sqlalchemy import create_engine
import sqlite3
```

```python
engine= create_engine('sqlite:///../../dataset/us/us_ticker_price_ma.db')
conn=sqlite3.connect('../../dataset/us/us_ticker_price_ma.db')
cursor = conn.cursor()
```

```python
#获取表名，保存在tab_name列表
cursor.execute("select name from sqlite_master where type='table'")
tab_name=cursor.fetchall()
tab_name_list = [line[0] for line in tab_name]
```

```python
len(tab_name_list)
```

```python
nice_list = []
error_list = []
for ticker in tab_name_list:
    try:
        df = pd.read_sql('{}'.format(ticker), engine)
        # 选取日期
        temp = df[df['datetime'] == '2022-07-11']
        # 条件
        # price>20MA
        # 10MA>50MA
        # 20MA>200MA
        # 50MA>200MA
        if temp.MA50.values[0] > temp.MA200.values[0] and temp.close.values[0] > temp.MA20.values[0] and temp.MA10.values[0] > temp.MA50.values[0]:
            nice_list.append(ticker)
    # df.rename(columns={'hgih':'high'},inplace=True)
    except:
        error_list.append(ticker)
        print('{} is error'.format(ticker))
```
