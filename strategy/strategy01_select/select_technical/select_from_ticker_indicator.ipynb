```python
import pandas as pd 
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///../../dataset/us/us_ticker_with_indicator.db')
df = pd.read_sql('2022-08-26',con=engine)
```

```python
df.rename(columns={'52week_low':'one_year_low',
                '52week_high':'one_year_high'},inplace=True)
df
```

```python
# # selection string
# conditon1 = 'Close > MA20'
# condi
```

```python
df = df.query('Close > MA20 and Close > one_year_low*1.25 and Close> 12 and MA20 > MA50 > MA200')
df.index = range(len(df))
df['rs_rank'] = 0
for i in range(len(df)):
    val = df.iloc[i].rs
    df['rs_rank'].iloc[i] = int(round(df[df.rs < val].rs.count()  / len(df),2)*100)
res = df.sort_values(by='rs_rank', ascending=False)
ticker_list = res[res.rs_rank > 75].symbol.to_list()
```
