```python
import tushare as ts
from config import token
import talib as ta
```

```python
pro = ts.pro_api(token=token)

#获取富时中国50指数
df = pro.index_global(ts_code='SPX')
```

```python
df = df[::-1]
df.index = range(len(df))
```

```python
df
```

```python
df['MA20'] = ta.SMA(df.close, timeperiod=20)
```

```python
df
```

```python
spx_ma = (df.iloc[-1].MA20  - df.iloc[-365].MA20) / df.iloc[-365].MA20
```
