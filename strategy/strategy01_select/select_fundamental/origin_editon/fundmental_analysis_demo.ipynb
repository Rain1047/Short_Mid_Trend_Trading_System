```python
# 当季每股净收益
# 每股净收益 = （税后净收益 - 优先股股息）/ 发行在外普通股股数

# 税后净收益
# revenue:收入
# earnins:净收入
net_earnings = msft.quarterly_earnings
# 显示股息(每股)
dividends = msft.dividends
# 发行在外的股数
nums_of_shares = msft.info['sharesOutstanding']

# 计算eps
net_earnings['eps'] = net_earnings.Earnings.astype(float) / nums_of_shares
#
```

```python
net_earnings
```

```python
# 4季度平均每股收益
net_earnings['eps'].mean()
```

```python
# 当季每股收益和当年最高差值
max_eps = net_earnings.eps.max()
max_eps
currect_eps_diff = net_earnings.tail(1)['eps'].values[0] - max_eps
currect_eps_diff
```
