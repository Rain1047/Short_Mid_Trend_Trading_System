```python
import yfinance as yf
from sqlalchemy import create_engine
price_engine = create_engine('sqlite:///../../../../dataset/us/us_ticker_price_yf.db')
pd.set_option('mode.chained_assignment', None)

for res in res_list:
    ticker_list = res.ticker.to_list()
    error_ticker_list = []
    for ticker in ticker_list:
        try:
            price_df = pd.read_sql('{}'.format(ticker), con=price_engine)
            price = price_df.close.values[0]
        except:
            error_ticker_list.append(ticker)
            print('{} not exits'.format(ticker))

    # get the error list
    if len(error_ticker_list) > 1:
        ticker_str = ''
        for ticker in error_ticker_list:
            ticker_str = ticker_str + str(ticker) + ' '

        try:
            data = yf.download(
                tickers = ticker_str,
                period = '5y',
                interval = '1d',
                group_by = 'ticker',
                auto_adjust = True,
                treads = True
            )
        except:
            continue

        for ticker in error_ticker_list:
            try:
                df = data['{}'.format(ticker)]
                if df.isnull().values.any() == True:
                    df.rename(columns={'High':'high','Low':'low','Close':'close','Open':'open'},inplace=True)
                    df['datetime'] = df.index
                    df.to_sql('{}'.format(ticker), if_exists='replace', con=price_engine, index=None)
                    print('{} price loaded'.format(ticker))
                else:
                    print('data has nan')
            except:
                print('{} oops'.format(ticker))
    else:
        continue
```
