```python
import numpy as np 
import pandas as pd 
from sqlalchemy import create_engine
```

```python
price_engine = create_engine('sqlite:///../../dataset/us/us_ticker_seven_year_price.db')
df = pd.read_sql('AAPL',con=price_engine)
df
```

```python
df['chg'] = 0
```

```python
df.chg = (df.Close - df.Open) / df.Open
df
```

```python
df.iloc[7].High
```

```python
pd.set_option('mode.chained_assignment', None)
df['seven_day_ir'] = 0
for i in range(7,len(df)):
    # 7-day increase rate
    df['seven_day_ir'].iloc[i] = (df.iloc[i].High - df.iloc[i-7:i].Close.min()) / df.iloc[i-7:i].Close.min()
df[df['seven_day_ir'] > 0.2]
```

```python
df['15_day_id'] = 0
for i in range(10,len(df)):
    # 7-day increase rate
    temp = df.iloc[i-10:i]    
    df['15_day_id'].iloc[i] = len(temp[temp.chg > 0])
select = df[df['15_day_id'] >= 7]
select.tail(20)
```
