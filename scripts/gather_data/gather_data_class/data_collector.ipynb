```python
import pandas as pd
df = pd.DataFrame(
    columns=['p','q']
)
df
```

```python
x = [46,42,38,34,32,30]
y = [50,70,84,100,108,116]
df.p = x
df.q = y
df['t'] = df.p * df.q
df['diff_q'] = df.q.diff().fillna(1)
df['diff_t'] = df.t.diff().fillna(1)
df['mr'] = df['diff_t'] / df['diff_q']
df
```

```python
import time
```

```python
def print_ticker(ticker):
    print(ticker)
    time.sleep(0.01)
```

```python
from concurrent.futures import ThreadPoolExecutor
with ThreadPoolExecutor(8) as executor:
    executor.map(print_ticker,ticker_list)
```

```python
for ticker in ticker_list:
    print_ticker(ticker)
```
