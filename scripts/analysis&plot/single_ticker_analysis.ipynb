## 存储SSE

````python
# Select SZSE List
conn = sqlite3.connect('ticker_symbol_ch.db')
cur = conn.cursor()
sql = """select * from SZSE"""
cur.execute(sql)
rows = cur.fetchall()
cur.close()
conn.close()
pro = ts.pro_api()
````

````python
# 连接到sse数据库
conn = sqlite3.connect('./ch/sse.db')
# 获取rows的长度
len_ = len(rows)
for i in range(len_):
    # 获取单独的股票的历史记录
    df = pro.daily(ts_code = rows[i][0])
    # 创建表
    try:
        cursor = conn.cursor()
        cursor.execute('CREATE TABLE {}(ts_code, trade_date, open, high, low, close, change, pct_cha, vol)'.format(rows[i][0]))
        conn.commit()
        print("{} ceate successful!".format(rows[i][0]))
    except:
        print("Table name already exists")
    # 写入数据
    # 写入（如果没有就写入，否则建立资料表）
    try:
        df.to_sql('{}'.format(rows[i][0]), conn, if_exists='replace',index=False)
    except:
        print('df already exits')
    
````

```python
def get_df_from_db(sql):
    cursor = connection.cursor()
    cursor.execute(sql)
    data = cursor.fetchall()
    columnDes = cursor.description #获取连接对象的描述信息
    columnNames = [columnDes[i][0] for i in range(len(columnDes))]
    df = pd.DataFrame([list(i) for i in data],columns=columnNames)
    return df
```
